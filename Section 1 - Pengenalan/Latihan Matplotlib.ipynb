```python
# kita membuat basic line plot
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([1, 2, 3])
plt.plot(myarray)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# membuat scatter plot
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

plt.scatter(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
